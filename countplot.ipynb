# Countplot

**Main Ideas:**

-A countplot shows the number of observations for each category in your data.

-It works like a histogram, but instead of bins of numeric values, it uses categorical variables.

-Useful for exploratory data analysis because it helps you quickly see how many times each category appears.

```python
import seaborn as sns
```

```python
diamonds = sns.load_dataset('diamonds')

diamonds.head()
```

```python
diamonds = diamonds[diamonds.clarity.isin(['SI1', 'VS2'])]
```

# Basics

```python
sns.set_style('darkgrid')
```

```python
sns.countplot(x='color', data=diamonds, hue='color', dodge=False, legend=False);
```

**This counts how many diamonds belong to each color and shows them as bars.**

```python
sns.countplot(y='color', data=diamonds, hue='color', dodge=False, legend=False);
```

**You can flip orientation with y="color" instead of x.**

# Ordering

```python
diamonds.color.value_counts().index
```

```python
sns.countplot(x='color', data=diamonds,
              order=diamonds.color.value_counts().index[::-1],  #remove [::-1] for descending
             hue='color', dodge=False, legend=False); 
```

From the graph, we see:
Color E is the most frequent.
Color J is the least frequent.

Middle categories (D, H, G, F) are fairly balanced in frequency.
This visualization helps quickly identify which diamond colors dominate the dataset.

```python
sns.countplot(x='color', data=diamonds, hue='clarity', hue_order=['SI1', 'VS2']);
```

This countplot shows the distribution of diamond colors.

The bars are split by clarity (SI1 vs. VS2) using the hue argument.

The hue_order ensures the legend only includes the two filtered categories (SI1 and VS2).

# Styling

**To customize bar colors, we can use palette or color.**

```python
sns.countplot(x='color', data=diamonds, color='skyblue');
```

```python
sns.countplot(x='color', data=diamonds, palette='twilight', hue='color', dodge=False, legend=False);
```

# Styling Using Mathplotlib

```python
sns.countplot(x='color', data=diamonds,
              lw=4, ec='black', hatch='/', 
              hue='color', dodge=False, legend=False
             );
```

lw=4 and ec='black' add thick black edges around the bars for emphasis.

hatch='/' applies a pattern fill (diagonal slashes) to each bar, making it visually distinct.

```python
from matplotlib import pyplot as plt
```

```python
sns.set(style='darkgrid', font_scale=1.3)
sns.countplot(x='cut', data=diamonds, color='navy', lw=3)
plt.xlabel('')
plt.ylabel('Count')


plt.tight_layout()
#plt.savefig('20countplot_image_countplotExample.png', dpi=200);
```

Shows diamond count by cut (Fair, Good, Very Good, Premium, Ideal).
All bars are navy blue with thick white edges.
Uses Seaborn darkgrid style with larger font size for readability.  plt.savefig() can be used to export the figure at high resolution.

# Summary

- Seaborns countplot is a simple but powerful tool to visualize category frequencies.
- Great for quickly spotting distributions and differences in your dataset.
- Flexible with ordering, hue, and styling options.
